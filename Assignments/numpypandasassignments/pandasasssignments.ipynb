
### Assignment 4: Data Aggregation

1. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'. Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers. Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category.


```python
df4 = pd.DataFrame(
    {
        'Category':np.random.choice(['A','B','C'],size = 10),
        'Value':np.random.randint(1,10,size=10)
    }
)
df4
```

```python
# This is how we group the data
print(df4.groupby('Category')['Value'].value_counts(),'\n')

print(df4.groupby('Category')['Value'].sum(),"\n")
print(df4.groupby('Category')['Value'].mean())
```


### Assignment 5: Merging DataFrames

1. Create two Pandas DataFrames with a common column. Merge the DataFrames using the common column.


```python
df5a = pd.DataFrame({
    'C1':['A','B','C','D','E'],
    'values':[1,2,3,4,5]
})

df5b = pd.DataFrame({
    'C1':['A','B','C','D','E'],
    'values':[1,2,3,4,6]
})

print(df5a)
print("\n")
print(df5b)
```

```python
pd.merge(df5a,df5b,how='inner')
```

```python
pd.merge(df5a,df5b,how='outer')
```

```python
pd.merge(df5a,df5b,how='left')
```

```python
pd.merge(df5a,df5b,how='right')
```


### Assignment 6: Time Series Analysis

1. Create a Pandas DataFrame with a datetime index and one column filled with random integers. Resample the DataFrame to compute the monthly mean of the values.


```python
date_range = pd.date_range(start='1/1/2025',end='31/12/2025',freq='D')
df6 = pd.DataFrame(date_range,columns=['Date'])
df6
```

```python
df6.set_index('Date',inplace=True)
```

```python
df6['Data'] = np.random.randint(1,10,size=(len(date_range)))
df6
```

```python
# Downsampling the data, lowering the frequency to months
resampled = df6.resample("ME").mean()
resampled
```


2. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and one column filled with random integers. Compute the rolling mean with a window of 7 days.


```python
date_range = pd.date_range(start = '2021-01-01', end = '2021-12-31', freq='D')
df6 = pd.DataFrame(date_range,columns=['Date'])

df6['Data'] = np.random.randint(1,10,size=len(date_range))

df6.set_index('Date',inplace=True)

# Rolling mean : Rolling mean is the mean of certain number(window) of past values at that point.
rolling_mean = df6.rolling(window=7).mean() 
rolling_mean
```

```python
# Another example of rolling mean
df = pd.DataFrame([1,2,3,4,5],index=pd.date_range('2025-01-01','2025-01-05'),columns=['Data'])
df
```

```python
rolling_mean = df.rolling(window=3).mean()
rolling_mean

# The first two values have 'nan' values, cuz we dont have 3 values about it
```

1. Create a Pandas DataFrame with a MultiIndex (hierarchical index). Perform some basic indexing and slicing operations on the MultiIndex DataFrame.


```python
np.random.seed(10)
# Create a data structure
array = np.array([['A','A','B','B'],['A1','A2','B1','B2']])   

# Create multiIndices  (Can be done using multiple ways)
index = pd.MultiIndex.from_arrays(array, names = ['Category1','Category2'])

# Create Df
df8 = pd.DataFrame(np.random.randint(1,10,size = (4,4)),index=index,columns=['V1','V2','V3','V4'])

df8
```

```python
np.random.seed(10)
# Method2 
index = pd.MultiIndex.from_tuples(tuples=(('A','A1'),('A','A2'),('B','B1'),('B','B2')),names=['Category1','Category2'])

df8 = pd.DataFrame(np.random.randint(1,10,size=(4,4)),index=index,columns=['V1','V2','V3','V4'])

df8
```

```python
np.random.seed(10)
# Method 3
index = pd.MultiIndex.from_product([['A','B'],['A1','B1']],names=['Category1','Category2'])

df8 = pd.DataFrame(np.random.randint(1,10,size=(4,4)),index=index,columns=['V1','V2','V3','V4'])

df8
```

```python
np.random.seed(10)
# Method 4 (takes multiindex from dataframe)
data = pd.DataFrame({
    'Category1':['A','A','B','B'],
    'Category2':['A1','A2','B1','B2'],
    'V1' : np.random.randint(1,10,size=4),
    'V2' : np.random.randint(1,10,size=4),
    'V3' : np.random.randint(1,10,size=4),
    'V4' : np.random.randint(1,10,size=4)
})
index = pd.MultiIndex.from_frame(data[['Category1','Category2']])

data.set_index(index,inplace=True)
data.drop(columns=['Category1','Category2'],inplace=True)

data
```
