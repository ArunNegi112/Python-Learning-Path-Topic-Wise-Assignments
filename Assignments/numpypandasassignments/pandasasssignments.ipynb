1. Create a Pandas DataFrame with 4 columns and 6 rows filled with random integers. Set the index to be the first column.


```python
df1 = pd.DataFrame(np.random.randint(1,10,size = (6,4)),columns=['A','B','C','D'])
df1
```

```python
# Method 1
df1.index = [df1.loc[:,'A']]
df1
```

```python
# Set the index to be the first column and remove it from columns
'''This is because 'A' is a column while df1.loc[:,0] gives series of elements in 'A' column'''
df1 = pd.DataFrame(np.random.randint(1,10,size = (6,4)),columns=['A','B','C','D'])
df1.set_index('A', inplace=True)
df1

```


### Assignment 3: Data Cleaning

1. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers. Introduce some NaN values. Fill the NaN values with the mean of the respective columns.


```python
data = [np.random.choice([1,2,3,4,5,6,np.nan,np.nan,10,30,40]) for i in range(15)]
df3 = pd.DataFrame(np.array(data).reshape(5,3),index=['A','B','C','D','E'],columns=[1,2,3])
df3
```

```python
df3.fillna(df3.mean(),inplace=True)  # This works cuzz,
# df3.mean() returns a series of column wise means

df3
```


2. Create a Pandas DataFrame with 4 columns and 6 rows filled with random integers. Introduce some NaN values. Drop the rows with any NaN values.
